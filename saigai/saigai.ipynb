```python
import pandas as pd
import matplotlib.pyplot as plt

# CSVèª­ã¿è¾¼ã¿
df = pd.read_csv("emdat.csv")

# åˆ—åã‚’çµ±ä¸€
df = df.rename(columns={
    "Start Year": "Year",
    "Disaster Type": "Disaster Type",
    "Country": "Country",
    "Total Deaths": "Deaths"
})

# æ¬ æã®ã‚ã‚‹Yearã‚’é™¤å¤–
df = df.dropna(subset=["Year"])

# å¹´ã‚’æ•´æ•°ã«
df["Year"] = df["Year"].astype(int)

# ç½å®³ã‚¿ã‚¤ãƒ—ä»¶æ•°ãƒ©ãƒ³ã‚­ãƒ³ã‚°
type_counts = df.groupby("Disaster Type").size().sort_values(ascending=False)

# ä¸Šä½5ã®ã¿æŠ½å‡º
top5_types = type_counts.head(5).index
df_top5 = df[df["Disaster Type"].isin(top5_types)]

# ========================
# ã‚°ãƒ©ãƒ•â‘ ï¼šå…¨ç½å®³ã‚¿ã‚¤ãƒ—ã®ä¸Šä½5
# ========================
plt.figure(figsize=(8,5))
type_counts.head(5).plot(kind="bar")
plt.title("Top 5 Disaster Types (All Years)")
plt.xlabel("Disaster Type")
plt.ylabel("Number of Events")
plt.tight_layout()
plt.savefig("top5_disaster_types.png")
plt.close()

# ========================
# ã‚°ãƒ©ãƒ•â‘¡ï¼šä¸Šä½5ã‚¿ã‚¤ãƒ— Ã— å¹´åˆ¥æ¨ç§»
# ========================
pivot = df_top5.groupby(["Year", "Disaster Type"]).size().unstack(fill_value=0)
pivot.plot(kind="bar", stacked=True, figsize=(12,6))
plt.title("Yearly Trend of Top 5 Disaster Types")
plt.xlabel("Year")
plt.ylabel("Number of Events")
plt.tight_layout()
plt.savefig("yearly_trend_top5.png")
plt.close()

print("ğŸ“Œ å®Œäº†ï¼ç”Ÿæˆã•ã‚ŒãŸãƒ•ã‚¡ã‚¤ãƒ«ï¼š")
print("- top5_disaster_types.png")
print("- yearly_trend_top5.png")
```
